```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import pandas as pd
import numpy as np
```

```python
x, y = make_classification(n_samples=800, n_features=2, n_informative=2, n_redundant=0, random_state=25)
y=[-1 if (i==0) else 1 for i in y ]
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
set(y)
```

```python
from SVM_hinge import SVM_hinge
model=SVM_hinge()
model.fit(X_train,y_train,lring_rate=.01,num_iters=500,lambada=.7)
```

```python
y_pred=model.predict(X_train)
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

pd.DataFrame([y_pred,y_train])
```

```python
confusion = confusion_matrix(y_train,y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix=confusion, display_labels=np.unique(y))
cm_display.plot()
```
